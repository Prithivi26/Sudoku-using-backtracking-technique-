```python
def Display_board(Grid):
    for i in range(9):
        if i%3 == 0:
            print("-"*30)
        for j in range(9):
            if j%3 == 0:
                print(end="|")
            print(Grid[i][j], end=" ")
        print("|")
    print("-"*30,"\n")

def Find_Empty_Position(Grid):
    for i in range (9):
        for j in range(9):
            if(Grid[i][j] == 0):
                return i,j
    return False

def is_valid(Grid,Num,Position):
    for i in range(9):
        if(Grid[Position[0]][i] == Num):
            return False
    for i in range(9):
        if (Grid[i][Position[1]] == Num):
            return False
    x, y =(Position[1]//3)*3,(Position[0]//3)*3

    for i in range(3):
        for j in range(3):
            if(Grid[y+i][x+j] == Num):
                return False
    return True

def Solve(Grid):
    Position = Find_Empty_Position(Grid)
    if(Position == False):
        return True

    for Num in range(1,10):
        if (is_valid(Grid,Num,Position)):
            Grid[Position[0]][Position[1]] = Num
            if (Solve(Grid)):
                return True
            Grid[Position[0]][Position[1]] = 0
    return False 
        
Grid = [[0, 7, 0, 0, 5, 0, 0, 1, 0],
       [0, 5, 0, 7, 2, 8, 6, 0, 0],
       [2, 6, 0, 0, 0, 0, 7, 0, 0],
       [0, 0, 0, 0, 0, 6, 0, 0, 0],
       [5, 3, 0, 0, 0, 0, 0, 0, 7],
       [0, 8, 0, 0, 9, 0, 0, 4, 0],
       [6, 0, 3, 0, 0, 0, 0, 8, 1],
       [7, 0, 5, 3, 8, 0, 0, 0, 0],
       [0, 0, 0, 0, 0, 9, 3, 7, 0]]     

Display_board(Grid)
Solve(Grid)
Display_board(Grid)
```
